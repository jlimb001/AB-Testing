```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('real01.csv')
df.head()
```

```python
# Arranging the data in ascending order

sorted_column = ['client_id', 'visitor_id', 'process_step', 'date_time']

df_sorted = df.sort_values(by=sorted_column, ascending=True)
df_sorted.head()
```

```python
# Grouping the columns to track individual user journeys

grouped_df = df_sorted.groupby(['client_id', 'visitor_id','visit_id'])
grouped_df.head()
```

```python
#Ensuring the timestamp column is in the correct datetime format

#print(grouped_df['date_time'].dtype)

def convert_datetime(group):
    group['date_time'] = pd.to_datetime(group['date_time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
    return group

grouped_df = grouped_df.apply(convert_datetime)


# grouped_df['date_time'] = pd.to_datetime(grouped_df['date_time'], format ='%Y-%m-%d %H:%M:%S', errors='coerce')
```
